```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "TestCase": ["TC1", "TC2", "TC3", "TC4", "TC5"],
    "Module": ["Login", "Login", "Payment", "Payment", "Reports"],
    "Status": ["Passed", "Failed", "Passed", "Failed", "Passed"],
    "Duration": [12, 15, 20, 18, 25],
    "Defects": [0, 2, 1, 3, 0]
}

df = pd.DataFrame(data)
print(df)
```

Univariate Analysis

Show distribution of one column: Duration.

Print mean, median, std of Duration.

```python
df['Duration'].describe()
```

```python
plt.hist(df['Duration'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribution of Duration")
plt.xlabel("Duration")
plt.ylabel("Frequency")
plt.show()
```

```python
mean_duration = df['Duration'].mean()
median_duration = df['Duration'].median()
std_duration = df['Duration'].std()

print(f"Mean Duration: {mean_duration}")
print(f"Median Duration: {median_duration}")
print(f"Standard Deviation of Duration: {std_duration}")
```

Bivariate Analysis

Compare Status vs Duration. (e.g., average duration by status). 

df.groupby("Status")["Duration"].mean()


```python
avg_df = df.groupby("Status")["Duration"].mean()
print("Average Duration by Status:")
print(avg_df)

avg_df.plot(kind='bar', color=['red', 'green'], edgecolor='black', figsize=(6,4))
plt.title("Average Duration by Status")
plt.xlabel("Status")
plt.ylabel("Average Duration")
plt.show()
```

Multivariate Analysis

Compare Module, Status, and Defects.

Example: df.groupby(["Module","Status"])["Defects"].sum()

```python
defects_summary = df.groupby(["Module", "Status"])["Defects"].sum()
print("Sum of Defects by Module and Status:")
print(defects_summary)
```
