```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Week1": [1000, 2000, 3000, 5000, 7000, 8000, 6700],
    "Week2": [4000, 5000, 2500, 4500, 3500, 5000, 5600],
    "Week3": [5000, 6000, 4500, 3500, 2000, 6000, 4800],
    "Week4": [6000, 5000, 2900, 4500, 3500, 4500, 4500]
}

df = pd.DataFrame(data)

print(df)
```

```python
day = df["Day"]
week1 = df["Week1"]
week2 = df["Week2"]
week3 = df["Week3"]
week4 = df["Week4"]
```

```python
df.plot(kind="line", marker="o", title="Line Plot - Weekly Data")
plt.ylabel("Values")
plt.xlabel("Days")
plt.grid()
plt.show()
```

```python
df.plot(kind='bar', x='Day', y=['Week1', 'Week2', 'Week3', 'Week4'], title='Bar Plot')
plt.show()
```

```python
df[["Week1", "Week2", "Week3", "Week4"]].plot(kind='hist', bins=20, alpha=1)
plt.title("Histogram of Weekly Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend(title="Weeks")
plt.show()
```

```python

modules = ['AI', 'ML', 'Python', 'DataStructures']
teamsize = [135, 125, 125, 115]

df1 = pd.DataFrame({
    'modules': modules,
    'teamsize': teamsize
})
df1.plot(kind='pie', y='teamsize', labels=modules, autopct='%1.1f%%', startangle=140)
plt.legend(title="Modules", loc = "upper left", bbox_to_anchor=(1.3, 0.9))
plt.title("Team Size Distribution by Module")
plt.show()
```
